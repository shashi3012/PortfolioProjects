```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjust the configuration pf the plots that we will create

# Read in the data
df= pd.read_csv(r'C:\Users\Rahul\Downloads\movies.csv')
                    
```

```python
df.head() # lets see the data
```

```python
# Lets see if there is any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
# Data types for our columns
df.dtypes
```

```python
# Change data type of columns
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
df['yearcorrect'] = df['released'].astype(str).str[:4]

```

```python
df.head()
```

```python
df=df.sort_values(by=['gross'],inplace= False , ascending= False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df=df.sort_values(by=['gross'],inplace=False, ascending=False)
```

```python
# Drop any duplicates.

Unique_matrix = df.drop_duplicates()
Unique_matrix.head()

```

```python
# Budget high correlation
# Company high correlation

#scatter plot with budget versus gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Films')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
df.head()
```

```python
# Plot Budget vs Gross using Seaborn

sns.regplot(x='budget',y='gross', data = df, scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
#lets start looking at correlations

df.corr(method='pearson')
```

```python
#High Correlation Between Gross And Budget

correlation_matrix= df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot= True)
plt.title('Correlation Matrix For Numeric Feature')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name]= df_numerized[col_name].cat.codes
        
df_numerized.head()        
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```
